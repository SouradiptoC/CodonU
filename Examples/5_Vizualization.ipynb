# One Advanced Example

Often, you may need to compare the gene COA data. As of now i'm unable to provide a separate function which can take variable amount files and plot the gene COA data in one plot with necessary ligands. How ever, here I am showing how you can achieve it.

```python
# Importing necessary files
import matplotlib.pyplot as plt
from CodonU.correspondence_analysis import build_contingency_table_codon_count, ca_codon
```

```python
# defining inner params
genetic_code_num = 11

# defining separate input files
aures_in_file = 'Nucleotide/Staphylococcus_aureus.fasta'
coli_in_file = 'Nucleotide/e_coli.fasta'

# defining codons belonging to synonymous family 1 [SF1]
single_syn_codons = ['ATG', 'TGG']
```

```python
# Performing CA on aures
aures_cont_table = build_contingency_table_codon_count(aures_in_file, genetic_code_num)    # building contingency table
aures_cont_table.replace(0, 0.000001, inplace=True)    # replacing 0 with 0.000001 for accuracy in calculation
aures_ca = ca_codon(aures_cont_table, n_components=2, single_syn_codons=single_syn_codons)    # performing CA
aures_cont_table.drop(single_syn_codons, axis=1, inplace=True)    # dropping codons of SF1 from contingency table, as they are useless now
aures_cont_table = aures_cont_table.astype('float64')    # converting the data type so that no type compatibility issue arises
aures_genes = aures_ca.row_coordinates(aures_cont_table)    # finally taking the coordinates for genes
```

```python
# Performing CA on coli
coli_cont_table = build_contingency_table_codon_count(coli_in_file, genetic_code_num)
coli_cont_table.replace(0, 0.000001, inplace=True)
coli_ca = ca_codon(coli_cont_table, n_components=2, single_syn_codons=single_syn_codons)
coli_cont_table.drop(single_syn_codons, axis=1, inplace=True)
coli_cont_table = coli_cont_table.astype('float64')
coli_genes = coli_ca.row_coordinates(coli_cont_table)
```

```python
# Plotting
aures_x, aures_y = aures_genes.iloc[:, 0], aures_genes.iloc[:, 1]
coli_x, coli_y = coli_genes.iloc[:, 0], coli_genes.iloc[:, 1]
fig = plt.figure(figsize=(16, 9))
plt.scatter(aures_x, aures_y, alpha=1, marker='*', c='#5ab4ac', s=10, label='S aures', zorder=2)
plt.scatter(coli_x, coli_y, alpha=0.7, marker='x', c='#d8b365', s=10, label='E coli', zorder=2)
plt.grid(True, linestyle=':')
plt.axvline(0, color='red', zorder=1)
plt.axhline(0, color='red', zorder=1)
plt.xlabel('Axis 0')
plt.ylabel('Axis 1')
plt.legend()
plt.title('Gene COA Aures VS Coli')

plt.show()
plt.close(fig)
```
