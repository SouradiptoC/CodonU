# Graphics

## ENc Plot

```python
in_file = 'Nucleotide/Staphylococcus_agnetis_nucleotide.fasta'
viz.plot_enc(in_file, genetic_table_num=11, organism_name='Staphylococcus agnetis')
```

As you can see above, the figure has a scale of ENc value. There is a title, and total number of genes shown in the plot. There are some other arguments of the function, viz.
- `min_len_threshold`: Minimum length of nucleotide sequence to be considered as a gene. (Default 200)
- `folder_path`: Path to folder. (Default _Report_)
- `gene_analysis`: To show individual genes in the plot. If want the genome as whole, just pass `false`

If `organism_name` is not given, then the title will change, rest will same, as shown below. Thse second argument is `genetic_table_num` which has been discussed in Analyzer.

```python
viz.plot_enc(in_file, 11)
```

## Neutrality Plot

```python
viz.plot_neutrality(
    handle=in_file,
    min_len_threshold=200,
    organism_name='Staphylococcus agnetis'
)
```

In case of neutrality plot, we can see total number of genes as well as R^2 value. The scale on the right side will vary with the distribution of GC3 and GC12 value. In case of _Staphylococcus agnetis_, the scale is based on GC3 because, the value range for GC3 is more diverse compared to GC12. As shown below, for _Staphylococcus agnetis_, the scale is based on GC12 as its values are more diverse.

```python
viz.plot_neutrality(
    'Nucleotide/Staphylococcus_argenteus_nucleotide.fasta',
    min_len_threshold=200,
    organism_name='Staphylococcus argenteus'
)
```

## Parity Rule 2 Plot

```python
viz.plot_pr2(
    handle=in_file,
    min_len_threshold=200, 
    organism_name='Staphylococcus agnetis'
)
```

In the above polt, total number of genes in each quadrant and on the axis are given in bottom-left corner. A legend is also given in the top-right position.

You may have noticed the area of the dots are not same in case of correspondence analysis plots. The area of the dots are directly related to the length of the sequence. Higher the length of the sequence is, higher the area of the dots will be.
